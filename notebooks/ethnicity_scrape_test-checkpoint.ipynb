```python
import requests
from bs4 import BeautifulSoup  
import pandas as pd
import numpy as np
import re
```

```python
website_url = 'https://statisticalatlas.com/place/Tennessee/Belle-Meade/Race-and-Ethnicity'
response = requests.get(website_url)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.title)
```

```python
print(soup)
```

```python
bellemeade = soup.find_all('div', class_ = 'figure-contents')
print(type(bellemeade))
print(len(bellemeade))
```

```python
first_race = bellemeade[0]
first_race
```

```python
race_text = first_race.g.text
race_text
```

```python
race = str(race_text)
race_reg = re.findall(r"[W,H,B,A,M,O]{1}[a-z]+", race)
print(race_reg)
```
