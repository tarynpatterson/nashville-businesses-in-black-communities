The purpose of this notebook is to scrape the Nashville Black Chamber of Commerce website to get a list of Businesses and their addresses.

```python
import requests
from bs4 import BeautifulSoup  
import pandas as pd
```

```python
url = 'https://nashvilleblackchamberofcommerce.growthzoneapp.com/directory/FindStartsWith?term=%23%21'
response = requests.get(url)

response.status_code
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.title)
```

```python
print(soup)
```

```python
#business_names = soup.find('span', class_ = "gz-img-placeholder").text
#print(business_names)
```

```python
#declaring the list of empty variables, So that we can append the data overall

business_names = []

#the whole core of the script
for names in soup:
    names = soup.find_all('span', class_ = "gz-img-placeholder")
    for store in names:
        business = store.text
        business_names.append(business)
    
print(business_names)
```

```python
len(business_names)
```

```python
business_names_unique = []
for x in business_names:
    if x not in business_names_unique:
        business_names_unique.append(x)
print(business_names_unique)
```
