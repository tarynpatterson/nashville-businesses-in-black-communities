```python
import pandas as pd
import googlemaps
import pprint
import time
import folium
```

```python
hidden = pd.read_pickle('../data/API_key.pkl')
```

```python
API_key = hidden.loc[0][1]
```

```python
gmaps = googlemaps.Client(key=API_key)
```

```python
places_nashville = gmaps.places_nearby(location= '36.1627,-86.7816', radius= 40000)
```

```python
pprint.pprint(places_nashville)
```

```python
location = '36.1627,-86.7816'
radius = 40000
token = places_nashville['next_page_token']
```

```python
time.sleep(3)
nashville_businesses = gmaps.places_nearby(location=location, radius=radius, page_token=token)
pprint.pprint(nashville_businesses)
```

```python
for place in places_nashville['results']:
    my_place_id = place['place_id']
    my_fields = ['name', 'type']
    place_details = gmaps.place(place_id = my_place_id, fields = my_fields)
    print(place_details)
```

```python
nash_map = folium.Map([36.1612, -86.7775], zoom_start = 12)

for store in places_nashville['results']:
    lat_lng = store['geometry']['location']
    location = [lat_lng['lat'], lat_lng['lng']]
    popup = store['name']
    folium.Marker(location = location, popup = popup).add_to(nash_map)

nash_map
```
